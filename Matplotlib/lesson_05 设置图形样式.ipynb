```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 注意figured的尺寸单位是英寸
# dpi:分辨率; frameon:背景框架，默认True
plt.figure(figsize=(10, 5), facecolor='r', edgecolor='y', linewidth=5, dpi=200, frameon=True)
plt.plot(np.arange(5, 20))

# 保存图片(可以直接右键图片另存为，更好用)
plt.savefig("./x.png")

# 中间内容的设置需要用Axes，Axes举例：
# ax= plt.gca()
# ax.set_facecolor("g")
```

# 笔记：
1. 画板的样式：通过调用plt.figure来实现，这个函数调用必须要在所有绘图之前来完成。参数解释：
    * figsize：画板的尺寸，是一个元组，第一个参数是宽度，第二个参数是高度。单位是英寸。
    * facecolor：画板颜色，注意区分Axes的背景颜色。
    * edgecolor：边框颜色，边框默认的宽度是0，因此如果想要看到边框，那么需要设置linewidth=2。
    * dpi：像素，也就是一个英寸中的像素点，默认是100，如果想要增加分辨率，那么就可以把dpi设置得大一点。
2. 保存图片：
    * plt.savefig()。
    * 鼠标左键点击图片，然后右键保存图片。
