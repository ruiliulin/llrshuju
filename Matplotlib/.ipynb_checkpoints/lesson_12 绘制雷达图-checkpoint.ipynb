```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as f_m
import pandas as pd
import numpy as np
font = f_m.FontProperties(fname=r"C:\Windows\Fonts\msyh.ttc",size=12)
```

```python
# 使用plt.polar()绘制雷达图
properties = ['输出','KDA','发育','团战','生存']
# 为了闭合，数据要多写一个，并与头一个一致
values = [40, 91, 44, 90, 95, 40]

theta = np.linspace(0, 2*np.pi, 6)
plt.polar(theta, values)
plt.fill(theta, values)
plt.show()
```

```python
# 使用子图的方法绘制雷达图
# 1. plt.subplot()
axes = plt.subplot(111, projection="polar")
axes.plot(theta, values)
axes.fill(theta, values)
```

```python
# 使用子图的方法绘制雷达图
# 2. plt.subplots()
figure, axes = plt.subplots(1, 1, subplot_kw={'projection':"polar"})
axes.plot(theta, values)
```

```python
# 使用子图的方法绘制雷达图
# 3. fig.add_subplot()
fig = plt.figure(figsize=(10, 10))
axes = fig.add_subplot(111, polar=True)
axes.plot(theta, values)
```

# 总结：
1. 雷达图：用来清晰的表示多个维度的值的强弱关系。
2. 绘制雷达图：
    1. 因为雷达图绘制的时候，不会自动的封闭，所以要在数据的最后多添加一个起始点的坐标。
    2. plt.polar绘制。
    3. plt.subplot(111,projection="polar")。
    4. figure,axes = plt.subplots(1,1,subplot_kw={"projection":"polar"})
    5. axes = fig.add_subplot(111,polar=True)
