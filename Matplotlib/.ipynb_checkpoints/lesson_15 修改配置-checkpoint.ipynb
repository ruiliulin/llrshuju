```python
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
plt.plot(range(0, 10))
plt.plot(range(0, 10, 2))
```

```python
# 更改配置文件
from matplotlib import cycler
plt.rcParams['axes.prop_cycle'] = cycler('color', ['k', 'y']) # 更改线条默认配置颜色
# 设置默认字体（可显示中文）
plt.plot(range(0, 10))
plt.plot(range(0, 10, 2))
plt.rcParams["font.sans-serif"] = ["SimHei"] # 字体样式
plt.rcParams["font.size"] = 20 # 字体大小
plt.title("中文显示")
```

**有时候我们可能需要设置一大堆参数，并且这个配置在后面很多项目中可能都会用到，那么这时候我们就可以把这些配置信息放到文件中（可配置项见下），文件的命名规则为 [名称].mplstyle ，然后把这个文件放到 matplotlib.get_configdir()/stylelib 的目录中，在写代码的时候根据名称加载这个配置文件，示例代码如下：**

```python
# 自定义配置文件
# 配置文件存放路径
mpl.get_configdir()
```

```python
# 使用自定义配置文件
plt.style.use("peizhi")

plt.plot(range(0, 10))
plt.title("测试")
```

# 笔记:
1. 动态设置配置:通过plt.rcParams来设置.
2. 配置文件配置:把配置项放到[名称].pltstyle文件中,然后把这个文件放在`matplotlib.get_configdir()/stylelib`下,以后在代码中就通过`plt.style.use("名称")`就可以使用了.
3. 配置项: https://raw.githubusercontent.com/matplotlib/matplotlib/master/matplotlibrc.template
