```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib.font_manager as f_m
```

```python
font = f_m.FontProperties(fname="C:\Windows\Fonts\msyh.ttc", size=10)
```

```python
# 绘制条形图
# 票房数据，单位：亿元
movies = {
    "流浪地球":40.78,
    "飞驰人生":15.77,
    "疯狂的外星人":20.83,
    "新喜剧之王":6.10,
    "廉政风云":1.10,
    "神探蒲松龄":1.49,
    "小猪佩奇过大年":1.22,
    "熊出没·原始时代":6.71
}

x = list(movies.keys())
y = list(movies.values())
plt.figure(figsize=(15, 5))
# plt.bar常见参数(x, y, width：线条宽度, align:对齐方式， bottom: 底部起始位置 ......)
# plt.bar(x, y, width=-0.5, align="edge", color="r", edgecolor="b")
movies_df = pd.DataFrame({"names":list(movies.keys()), "tickets":list(movies.values())})
plt.bar('names', 'tickets', data=movies_df)
plt.xticks(fontproperties=font)
plt.yticks(range(0, 45, 5), ["%d亿" % x for x in range(0, 45, 5)], fontproperties=font)
```

```python
# 横向条形图
# 就是把y参数作为横轴， 把x参数作为纵轴，别的基本不变
plt.barh(list(movies.keys()), list(movies.values()))
plt.yticks(fontproperties=font)
```

```python
# 分组条形图
# 数据如下：
movies = {
    "流浪地球":[2.01,4.59,7.99,11.83,16],
    "飞驰人生":[3.19,5.08,6.73,8.10,9.35],
    "疯狂的外星人":[4.07,6.92,9.30,11.29,13.03],
    "新喜剧之王":[2.72,3.79,4.45,4.83,5.11],
    "廉政风云":[0.56,0.74,0.83,0.88,0.92],
    "神探蒲松龄":[0.66,0.95,1.10,1.17,1.23],
    "小猪佩奇过大年":[0.58,0.81,0.94,1.01,1.07],
    "熊出没·原始时代":[1.13,1.96,2.73,3.42,4.05]
}
movie_df = pd.DataFrame(movies)
movie_df
```

```python
plt.figure(figsize=(20, 5))
# for i in movie_df.index:
#     series = movie_df.iloc[i]
#     print(series)
#     plt.bar(series.index, series.values)

# 根据所给数据定义x轴
xticks = np.arange(0, len(movies), 1)
# 自定义线条的宽度，为画多个线条做基础
bar_width = 0.15

# 具体绘制方法
# plt.bar(xticks-2*bar_width, movie_df.iloc[0], width=bar_width)
# plt.bar(xticks-bar_width, movie_df.iloc[1], width=bar_width)
# plt.bar(xticks, movie_df.iloc[2], width=bar_width)
# plt.bar(xticks+bar_width, movie_df.iloc[3], width=bar_width)
# plt.bar(xticks+2*bar_width, movie_df.iloc[4], width=bar_width)

# 将以上画图算法进行循环绘制计算
for i in movie_df.index:
    
    x = xticks+(-2+i)*bar_width
    y = movie_df.iloc[i]
    # label是标签，和下面的plt.legend()配合使用
    plt.bar(x, y, width=bar_width, label="第%d天票房" % (i+1))
    # 添加文本注释
    for k, v in zip(x, y):
        plt.annotate("%.1f" % v, xy=(k, v), xytext=(k-0.08, v+0.2))
        
# 修改轴刻度,标题等
plt.xticks(xticks, movie_df.columns, fontproperties=font, size=15)
plt.yticks(range(0, 20, 2), range(0, 20, 2), fontproperties=font, size=15)
plt.ylabel("票房：亿", fontproperties=font, size=15)
plt.title("春节前5天电影票房记录", fontproperties=font, size=20)

# 添加图例
font.set_size(12)
plt.legend(prop=font)

plt.grid()
plt.show()
```

```python
# 堆叠条形图
# 数据
menMeans = (20, 35, 30, 35, 27)
womenMeans = (25, 32, 34, 20, 25)
groupNames = ('G1','G2','G3','G4','G5')

plt.figure(figsize=(10, 5))
# 使用bottom属性
plt.bar(groupNames, menMeans, label="男性得分")
plt.bar(groupNames, womenMeans, bottom=menMeans, label="女性得分")

plt.legend(prop=font)
```

# 笔记：
1. 绘制条形图使用的是plt.bar方法。这个方法有很多参数：
    * x：在x轴上的坐标，可以为一个列表/数组/字符串。
    * y：在y轴上的坐标，可以为一个列表/数组/字符串。
    * data：如果传递了data，那么x和y就可以为data中的key。比如data是一个DataFrame对象，那么x和y就是这个DataFrame对象的某个列的名字。
2. 官方文档：https://matplotlib.org/api/_as_gen/matplotlib.pyplot.bar.html#matplotlib.pyplot.bar
3. 横向条形图：plt.barh。然后y代表的是之前bar的x，width代表的是bar的height。其他参数都差不多。
4. 分组条形图：matplotlib并没有提供专门的分组条形图的API来实现。而是需要设置这个坐标来实现。绘制的规则是，一次性绘制所有电影第1天的数据，第二天的数据2...，然后要找到规律，把这个条形图的位置算好。
5. 堆叠条形图：matplotlib并没有提供专门的堆叠条形图的API来实现，我们在绘制往上堆叠的数据的时候，可以使用bottom参数，来往上偏移指定的距离。这样就可以实现堆叠的效果了。
