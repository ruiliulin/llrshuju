```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 导入字体设置
from matplotlib import font_manager
```

```python
y = [np.random.randint(10) for x in range(10)]
# 设置字体（可以显示中文）
font = font_manager.FontProperties(fname=r"C:\Windows\Fonts\msyh.ttc", size=20)
plt.title('折线图',fontproperties=font)

plt.plot(y)
```

```python
avenger = [17974.4, 50918.4, 30033.0, 40329.1, 52330.2, 19833.3, 11902.0, 24322.6, 47521.8, 32262.0, 
           22841.9, 12938.7, 4835.1, 3118.1, 2570.9, 2267.9, 1902.8, 2548.9, 5046.6, 3600.8]
# 设置图形规模尺寸
plt.figure(figsize=(15, 5))
# 设置标记
plt.plot(avenger, marker='o')

# 改变字体大小
font.set_size(10)
# 添加x轴的刻度，注意要配合plt.show()使用
plt.xticks(range(20), ['第%d天' % x for x in range(1,21)], fontproperties=font)
# 设置x轴标签
plt.xlabel("天数", fontproperties=font)
# 设置y轴标签
plt.ylabel("票房（单位：万）", fontproperties=font)
# 设置标题
font.set_size(20)
plt.title("复仇者联盟上映后前20天的票房数据统计", fontproperties=font)
# 设置网格
plt.grid(True)
# 展示出来
plt.show()
```

# 笔记：
- 1.标题是通过plt.title来设置的。
- 2.刻度是通过xticks和yticks来设置的。xticks和yticks需要传递两个参数，第一个参数ticks不重要，只要跟数据量保持一致就可以了，比如有20个数据，那么只要产生20个数据的数组或者列表都可以，第二个参数labels是用来设置每个刻度上的文本显示，这个就比较重了。
- 3.轴的名称是通过xlabel和ylabel来设置的。
- 4.默认plot是不支持中文的，如果想要支持中文，那么应该创建一个字体对象matplotlib.font_manager.FontProperties，然后指定fname参数，需要具体的路径。这里还有一个小细节，拷贝出来的字体的字符串，前面有一个乱码，要把那个乱码删掉。获取字体路径的方式：右键->属性->安全->对象名称
