```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 传递线条参数来设置
y = [np.random.randint(10) for x in range(10)]
plt.plot(y, lw=4, c='g', ls='--', aa=True)
```

```python
# 对多个线条的其中一条进行设置
y1 = [np.random.randint(10) for x in range(10)]
y2 = [np.random.randint(10) for x in range(10)]
# lines接收到两条线
lines = plt.plot(range(10), y1, range(10), y2)
# 取出第一条线
line = lines[0]
# 对其进行设置
line.set_c('k')
line.set_linewidth(5)
line.set_alpha(0.5)

print(lines)
```

```python
# 对多个线条进行统一设置(用的少)
y1 = [np.random.randint(10) for x in range(10)]
y2 = [np.random.randint(10) for x in range(10)]
lines = plt.plot(range(10), y1, range(10), y2)

plt.setp(lines, c='g', alpha=0.5, lw=3, ls='--')
print(lines)
```

# 笔记：
- 1.使用plot绘图的时候，就可以传递Line2D的属性值进去，来修改线条的样式。
- 2.使用plot返回的线条，单独去设置也可以，但是单独设置的时候，是通过set_Line2D属性名()的方式来设置的。
- 3.使用plt.setp来一次性设置多个线条的样式。
