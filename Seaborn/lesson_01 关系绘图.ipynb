```python
import seaborn as sns
import pandas as pd
```

```python
tips = pd.read_csv("dataset/tips.csv")
tips.head()
```

```python
# sns.relplot(kind='scatter') 关系绘图
# 绘制散点图
sns.relplot(x="total_bill", y="tip", data=tips)
```

```python
#  添加hue,size参数
sns.relplot(x="total_bill", y="tip", data=tips, hue="day", size="size")
```

```python
# 添加col和row参数
sns.relplot(x="total_bill", y="tip", data=tips, row="time", col="sex")
```

```python
# 指定具体的列(换行)
sns.relplot(x="total_bill", y="tip", data=tips, col="day", col_wrap=3)
```

```python
# sns.scatterplot(x="total_bill", y="tip", data=tips)
# sns.relplot(x="total_bill", y="tip", data=tips, kind="scatter")
# 以上两行代码基本是一个意思，只不过relplot绘制散点图时 kind= 可以省略
```

```python
fmri = pd.read_csv("dataset/fmri.csv")
fmri.head()
```

```python
# 绘制折线图(置信区间可以不设置， ==> ci=None)
sns.relplot(x="timepoint", y="signal", data=fmri, kind="line")
```

```python
# 其他参数设置
sns.relplot(x="timepoint", y="signal", data=fmri, kind="line", hue="region", col="event", style="region")
```

```python
# 使用sns.lineplot() 绘制折线图
# lineplot 没有col,row,kind属性
sns.lineplot(x="timepoint", y="signal", data=fmri, hue="region",  style="region")
```

```python
# 也可以使用简单绘图
import numpy as np
sns.relplot(x="x", y="y", data=pd.DataFrame({"x":np.arange(10),"y":np.sin(np.arange(10))}), kind="line")
```

# 笔记：
1. relplot是用来绘制关系数据图的。scatterplot和lineplot是relplot的一个封装。
2. relplot在绘制的时候，不能直接给x和y指定具体的值，而应该使用data参数DataFrame中的列的名字。
3. hue表示的是颜色，应该指定为某个列的名字。那么sns会自动的将指定的列的值的个数，取不同的颜色。
4. col表示分成几个图像，应该指定某个列的名字。那么sns会自动的将DataFrame中这个列的数据，分成多个图。
5. style在绘制折线的时候，可以用来指定线条的样式。
