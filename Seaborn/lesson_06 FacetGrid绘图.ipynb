```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
tips = sns.load_dataset("tips")
```

```python
# 使用FacetGrid()绘图
g1 = sns.FacetGrid(tips)
# 使用map()配置
g1.map(plt.scatter,"total_bill","tip")
```

```python
# 传参（绘制多图并区分）
# g2 = sns.FacetGrid(tips, col="day", row="time")
g2 = sns.FacetGrid(tips, col="day", hue="time")
g2.map(plt.scatter, "total_bill", "tip")
g2.add_legend()
```

```python
# height 图形尺寸，aspect 宽高比 
g3 = sns.FacetGrid(tips, col="smoker", height=4, aspect=1.5)
g3.map(sns.barplot, "day", "total_bill")
```

```python
# 设置字体
sns.set(rc={"font.sans-serif":"SimHei"})
g2 = sns.FacetGrid(tips, col="day", hue="time")
g2.map(plt.scatter, "total_bill", "tip")

# 设置图例
# g2.add_legend(title="标题", labels=["午餐", "晚餐"], label_order=["Lunch", "Dinner"] )
g2.add_legend(title="标题", label_order=["Lunch", "Dinner"] )
new_labels = ['午餐', '晚餐']
for t, l in zip(g2._legend.texts, new_labels):
    t.set_text(l)
```

```python
g = sns.FacetGrid(tips, col="day", row="time")
g.map(sns.regplot, "total_bill", "tip")
# 设置多个图的标题
g.set_titles(template="时间{row_name}/星期{col_name}")
```

```python
# 设置坐标轴属性
g = sns.FacetGrid(tips, col="day", row="time")
g.map(sns.scatterplot, "total_bill", "tip")

# 设置 x,y 轴的标签
g.set_axis_labels("账单", "小费")
# 设置背景颜色
g.set(facecolor="y")
# 设置 x，y 轴的刻度属性
g.set(xticks=range(0, 50, 10), xticklabels=["$%d" % x for x in range(0, 50, 10)])
```

```python
# 设置fig属性
g = sns.FacetGrid(tips, col="day", row="time")
g.map(sns.scatterplot, "total_bill", "tip")

g.fig.set_dpi(200)
```
