```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
# 绘制线性回归图
# lmplot()
# x,y 必须传字符串，data= 必须要
sns.lmplot(x="total_bill", y="tip", data=tips)
```

```python
# regplot()
# data= 可以不用
sns.regplot(x=tips["total_bill"], y=tips["tip"])
```

# 笔记:
1. regplot和lmplot都可以绘制线性回归图。
2. regplot的x和y可以为numpy的数组或者是Series对象。
3. 一些其他的参数，比如order，robust等，可以调节线性的准确度。
