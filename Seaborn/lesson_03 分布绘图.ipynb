```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
# 数据
t = pd.read_csv("dataset/titanic.csv")
t.head()
```

```python
# 绘制 分布直方图（不仅仅是直方图！！！）
# 开启颜色样式
sns.set(color_codes=True)
# 选择非nan数据
t_age = t[~np.isnan(t['age'])]

# 默认绘制kde曲线，可以设置不显示
# bins 分割数量
# rug 是否展示胡须下线，胡须越密集，说明数量越多越集中
# hist 也可以设置不显示
sns.distplot(t_age["age"], kde=True, bins=30, rug=True, hist=False)
```

```python
# 二变量绘图
# 数据
tips = pd.read_csv("dataset/tips.csv")

# 绘制分布组合图
# kind="reg" 可以设置回归绘图和核密度曲线
sns.jointplot(x="total_bill", y="tip", data=tips, kind="reg" )
```

```python
# 六边形图
# 数据
ath = pd.read_csv("athlete_events.csv")
ath_c = ath[ath["NOC"] == "CHN"]

# kind="hex" 六边形显示
# 六边形有一个区间大小，之前这些点落在这个六边形中越多颜色越深
# gridsize=50 分割六边形块数=50块
# height 尺寸 长高一起设置
# ratio 主图和附图的尺寸比
# space 主图和附图的间距
# marginal_kws 设置附图的属性，如 bins, rug,kde 等
sns.jointplot(x="Height", y="Weight", data=ath_c, kind="hex", gridsize=50, color="r", height=10, ratio=4, space=0.2,
             marginal_kws={"bins":50, "rug":True, "kde":True})
```

```python
# 成对绘图
# 数据
iris = pd.read_csv("dataset/iris.csv")
iris.head()
```

```python
sns.pairplot(iris)
```

```python
# vars=['', '', ...] 选择字段绘制
sns.pairplot(tips, vars=["total_bill", "tip"])
```

```python
# diag_kind：更改对角线的图形样式，auto, hist, kde。
# kind：scatter, reg。
sns.pairplot(iris, vars=["sepal_length", "sepal_width", "petal_length"], kind="reg", diag_kind="kde")
```

# 分布绘图：
1. 单一变量绘图：用的是`distplot`，这个函数不仅仅可以绘制直方图，还可以绘制KDE曲线以及rug线，rug线越集中，说明数据越集中。
2. 多变量绘图：用的是`jointplot`，这个函数比传统的散点图可以展示更多的信息，他是在顶部和右边展示两个直方图。
3. 成对绘图（pairplot）：用的是`pairplot`，可以通过`vars`参数用来控制只需要绘制数据集中的某几个字段。斜对角的图，表示的是数据集中，单一列的分布情况。
