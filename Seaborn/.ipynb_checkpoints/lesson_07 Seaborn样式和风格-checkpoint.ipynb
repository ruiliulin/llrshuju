```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
tips = sns.load_dataset("tips")
fmri = sns.load_dataset("fmri")
fmri.head()
```

```python
# 五种风格
# sns.set_style("white")
# sns.set_style("whitegrid")
# sns.set_style("dark")
# sns.set_style("darkgrid")
# sns.set_style("ticks")
axes = sns.scatterplot(x="total_bill", y="tip", data=tips)
```

```python
# 查看可设置的样式
sns.axes_style()
```

```python
# 使用with代码块设置样式！！！！
# 此方法必须用with执行，否则无法更改样式
with sns.axes_style("darkgrid", rc={"axes.facecolor":"y", "grid.color":"k"}):
    sns.lineplot(x="timepoint", y="signal", data=fmri)
```

```python
# sns.set() 最强的设置方法，全能，有一些默认设置！！！
# sns.set(context='notebook', style='darkgrid', palette='deep', font='sans-serif', font_scale=1, color_codes=True, rc=None)
sns.set(rc={"lines.linewidth":5, "font.sans-serif":"SimHei"})
sns.lineplot(x="timepoint", y="signal", data=fmri)
```
